```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Uber_data = pd.read_csv(r"C:\Users\mohda\Downloads\Uber Request Data By Labmentix 1.csv")
```

```python
Uber_data.head()
```

```python
Uber_data.columns = Uber_data.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")
```

```python
Uber_data.dropna(subset=['request_id','pickup_point','status','request_timestamp'],inplace=True)
```

```python
Uber_data=Uber_data.rename(columns={'incorrect_name':'time_stamp'})
```

```python
print(Uber_data.head())
```

```python
Uber_data['request_timestamp'] = pd.to_datetime(Uber_data['request_timestamp'], dayfirst=True, errors='coerce')
```

```python
# Extract hour and date
Uber_data['hour'] = Uber_data['request_timestamp'].dt.hour
Uber_data['day'] = Uber_data['request_timestamp'].dt.date
```

```python
sns.countplot(x='status',data=Uber_data)
plt.title("status distribution")
plt.show()
```

```python
sns.histplot(data=Uber_data,x='hour',hue='status',multiple='stack')
plt.title("Hourly Request By Status")
plt.show()
```

```python
sns.countplot(x='pickup_point',hue='status',data=Uber_data)
plt.title("Pickup point vs status")
plt.show()
```
